# Finding Unique Values From Genre 

```python
data.columns
```

```python
data['Genre']
```

```python
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
list1   #this is 2d list 
```

# Converting 2d list in 1d 

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
one_d
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
uni_list
```
